```python
import requests
from io import BytesIO
from zipfile import ZipFile
import geopandas as gpd
from config import DATA_DIR, STREETS_PATH
from pathlib import Path
```

```python
url = (
    "https://www12.statcan.gc.ca/census-recensement/2011"
    "/geo/RNF-FRR/files-fichiers/lrnf000r24a_e.zip"
)
```

```python
response = requests.get(url)
```

```python
with ZipFile(BytesIO(response.content)) as zip_file:
    zip_file.extractall(DATA_DIR / 'canada_streets')

    for file in zip_file.namelist():
        file_path = DATA_DIR / 'canada_streets' / file
        file_path.rename(file_path.parent / f'canada_streets{file_path.suffix}')
```
